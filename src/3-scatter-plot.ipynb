### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

def generate_scatter_plot_with_annotations():
    x_values = [1, 2, 3, 4, 5]
    y_values = [2, 4, 6, 8, 10]
    plt.figure(figsize=(8, 6)) 
    plt.scatter(x_values, y_values, color='#3498db', s=100, edgecolors='black', zorder=2) 
    plt.title('Scatter Plot with Annotations', fontsize=16)
    plt.xlabel('X Value', fontsize=12)
    plt.ylabel('Y Value', fontsize=12)
    for i, (x, y) in enumerate(zip(x_values, y_values)):
        # Adjust text position slightly to avoid overlapping the marker
        plt.annotate(f'({x}, {y})', (x, y), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9, color='#555555')

    # Add padding to the plot
    plt.tight_layout()

    # Display the plot
    plt.show()

# Call the function to generate and display the plot
if __name__ == "__main__":
    generate_scatter_plot_with_annotations()
```
